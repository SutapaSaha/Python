# Aggregation

```python
X = pd.DataFrame(np.random.rand(30).reshape(5,6),
                index = [['a','a','b','b','c'],[1,2,2,3,3]],
                columns = [["col1", "col1", "col2", "col2", "col3", "col3"], ["c1", "c2", "c3", "c4", "c5", "c6"]])
print(X)
```

```python
X.index.names = ['Row1', 'Row2']
X.columns.names = ['ColName1', 'ColName2']
X
```

```python
print(X.sum(), X.sum(axis = 1), sep ="\n\n")
```

```python
print(X, X.sum(level ='Row1'), X.sum(level = 'Row2'), sep ="\n\n")
```

```python
print(X, X.sum(level ='ColName1', axis = 1), X.sum(level = 'ColName2', axis = 1), sep ="\n\n")
```

```python
print(X, X.mean(level ='ColName1', axis = 1), X.mean(level = 'ColName2', axis = 1), sep ="\n\n")
```

```python
print(X, X.mean(level ='Row1'), X.mean(level = 'Row2'), sep ="\n\n")
```

```python
print(X, X.std(level ='Row1'), X.std(level = 'Row2'), sep ="\n\n")
```

```python
print(X, X.std(level ='ColName1', axis = 1), X.std(level = 'ColName2', axis = 1), sep ="\n\n")
```

# Merging Dataframe

```python
df1 = pd.DataFrame({'Key':['KA', 'KA', 'AP', 'MP', 'KA'], 'data1':range(5)})
df2 = pd.DataFrame({'Key':['KA', 'AP', 'UP'], 'data2':range(3)})
print(df1, df2, sep="\n\n")
```

```python
pd.merge(df1,df2)
```

```python
df1 = pd.DataFrame({'Key1':['KA', 'KA', 'AP', 'MP', 'KA'], 'data1':range(5)})
df2 = pd.DataFrame({'Key2':['KA', 'AP', 'UP'], 'data2':range(3)})
print(df1, df2, sep="\n\n")
```

```python
pd.merge(df1, df2, left_on='Key1', right_on='Key2')
```

```python
df1 = pd.DataFrame({'Key':['KA', 'KA', 'AP', 'MP', 'KA'], 'data1':range(5)})
df2 = pd.DataFrame({'Key':['KA', 'AP', 'UP'], 'data2':range(3)})
print(df1, df2, sep="\n\n")
```

```python
print(df1, df2, pd.merge(df1, df2, how="inner"),pd.merge(df1, df2, how="outer"), sep="\n\n" )
```

```python
print(df1, df2, pd.merge(df1, df2, how="left"), pd.merge(df1, df2, how="right"), sep="\n\n")
```

# Concatenation

```python
df1 = pd.DataFrame({'col1': ['AP', 'KA', 'KA', 'MH'], 'col2': [1, 1, 2, 3], 'lval': [1, 2, 3, 4]})
df2 = pd.DataFrame({'col1': ['AP', 'KA', 'TN', 'UP', 'MP'], 'col2': [1, 1, 2, 3, 6], 'lval': [4, 5, 6, 7, 9]})
print(df1, df2, sep="\n\n")
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis = 1)
```

```python
df1 = pd.DataFrame({'col1': ['AP', 'KA', 'KA', 'MH'], 'col2': [1, 1, 2, 3], 'lval': [1, 2, 3, 4]})
df2 = pd.DataFrame({'col1': ['AP', 'KA', 'TN', 'UP', 'MP'], 'col2': [1, 1, 2, 3, 6], 'rval': [4, 5, 6, 7, 9]})
print(df1, df2, sep="\n\n")
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis = 1)
```

```python
pd.concat([df1, df2], keys = ['x', 'y'])
```

```python
pd.concat([df1, df2], keys = ['x', 'y'], axis = 1)
```

```python
df3 = df1.set_index("col1")
df4 = df2.set_index("col1")
print(df3,df4, sep="\n\n")
```

```python
pd.concat([df3, df4], ignore_index = True)
```

# Combine Dataset(combine_first)

```python
df1 = pd.DataFrame({'a':[1.,np.nan,5.,np.nan], 'b':[np.nan, 2., np.nan, 6.], 'c': range(2, 18, 4)})
df2 = pd.DataFrame({'a':[5.,4.,np.nan, 3., 7.], 'b':[np.nan,3., 4., 6., 8.]})
print(df1, df2, sep ="\n\n")
```

```python
df3 = df1.combine_first(df2)
df3
```

```python
df4 = df2.combine_first(df1)
df4
```

# Melt Function

```python
df1 = pd.DataFrame({'Name': ['Student1', 'Student1', 'Student2', 'Student3', 'Student4'],
                    'Subject': ['Maths', 'English', 'English', 'Physics', 'Chemistry'],
                    'Marks': [98, 97, 96, 95, 96],
                    'Optional Subject': ['Physics', 'Chemistry', 'Maths', 'Chemistry', 'Physics']})
print(df1)
```

```python
df1.melt('Name')
```

```python
pd.melt(df1, id_vars = 'Name', value_vars = ['Subject', 'Marks'])
```
