```python
hdfStore1 = pd.HDFStore("./hdfex1.h5")
```

```python
X = pd.Series([10, 20, 30, 40, 50])
X
```

```python
hdfStore1["SeriesObj"] = X
hdfStore1["Stocks"] = Y1
```

```python
Y = hdfStore1["SeriesObj"]
Y
```

```python
Y = hdfStore1["Stocks"]
Y
```

```python
hdfStore1.put("Stocks.New", Y1, format = "table")
hdfStore1.select("Stocks.New")
```
