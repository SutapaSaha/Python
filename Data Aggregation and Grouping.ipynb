# Data Aggregation Funtion(AGG Function)

```python
df1 = {"Gender": ["M", "F", "M", "F", "M", "F", "M", "F", "M", "F"],
       "Group" : ["Group1", "Group2", "Group3", "Group1", "Group2", "Group3", "Group1", "Group2", "Group3", "Group1"],
       "CarTrip(KM)": [100, 200, 300, 400, 500, 50, 100, 1000, 1500, 1600],
       "AvgSizeofTrip": [1000, 2000, 3000, 2500,1500, 1000, 2000, 3000, 2500, 1500]  
      }
X = pd.DataFrame(df1, columns = ["Gender", "Group", "CarTrip(KM)", "AvgSizeofTrip"])
X
```

```python
X1 = X.groupby('Gender')
type(X1)
```

```python
X.groupby('Gender').agg(min)
```

```python
X.groupby('Gender').agg([min, max])
```

```python
X.groupby(['Gender', 'Group']).agg([min, max])
```

```python
def max_min(df):
    return df.max() - df.min()
```

```python
X.groupby(['Gender', 'Group']).agg([min, max, max_min])
```

```python
X.groupby(['Gender', 'Group']).agg([("Min",min),("Max", max),("Max_Min", max_min)])
```

# GroupBy with Apply

```python
df1 = {"Gender": ["M", "F", "M", "F", "M", "F", "M", "F", "M", "F"],
       "Group" : ["Group1", "Group2", "Group3", "Group1", "Group2", "Group3", "Group1", "Group2", "Group3", "Group1"],
       "CarTrip(KM)": [100, 200, 300, 400, 500, 50, 100, 1000, 1500, 1600],
       "AvgSizeofTrip": [1000, 2000, 3000, 2500,1500, 1000, 2000, 3000, 2500, 1500]  
      }
X = pd.DataFrame(df1, columns = ["Gender", "Group", "CarTrip(KM)", "AvgSizeofTrip"])
X
```

```python
X.groupby(['Gender', 'Group']).apply(min)
```

```python
X.groupby(['Gender', 'Group']).apply(max)
```

```python
def max_min(df):
    return df.max() - df.min()
X.groupby(['Gender', 'Group']).apply(max_min)
```

```python
def max_min(df, x):
    return df.max() - df.min() > 10
X.groupby(['Gender', 'Group']).apply(max_min, 10)
```

# Bucket Analysis

```python
X1.value_counts()
```

```python
X.groupby(X1).sum()
```

```python
X.groupby(X1).describe()
```

```python
X.groupby(X1).std()
```

# Pivot and Cross Tabulation

```python
df1 = {"Gender": ["M", "F", "M", "F", "M", "F", "M", "F", "M", "F"],
       "Group" : ["Group1", "Group2", "Group3", "Group1", "Group2", "Group3", "Group1", "Group2", "Group3", "Group1"],
       "CarTrip(KM)": [100, 200, 300, 400, 500, 50, 100, 1000, 1500, 1600],
       "AvgSizeofTrip": [1000, 2000, 3000, 2500,1500, 1000, 2000, 3000, 2500, 1500]  
      }
X = pd.DataFrame(df1, columns = ["Gender", "Group", "CarTrip(KM)", "AvgSizeofTrip"])
X
```

```python
X.pivot_table(index = "Gender", columns = "Group", values = "CarTrip(KM)")
```

```python
X.pivot_table(index = "Gender", columns = "Group")
```

```python
X.pivot_table(index = "Gender", columns = "Group", aggfunc = np.sum)
```

```python
X.pivot_table(index = "Gender", columns = "Group", aggfunc = np.std)
```

```python
X.pivot_table(index = "Gender", columns = "Group", aggfunc = np.var)
```

```python
X.pivot_table(index = "Gender", columns = "Group", aggfunc = {"AvgSizeofTrip" : np.sum, "CarTrip(KM)": np.mean})
```

```python
X.pivot_table(index = [ "Gender", "Group"], columns = "Group", aggfunc = {"AvgSizeofTrip" : np.sum, "CarTrip(KM)": np.mean})
```

```python
pd.crosstab(X.Gender, X.Group)
```

```python
pd.crosstab([X.Gender, X.Group], X.Group)
```
